```python
''' Implementation of log normal distribution '''
```

```python
import numpy as np
```

```python
mu, sigma = 3, 1 # mean and standard deviation
s = np.random.lognormal(mu, sigma, 10000)
```

```python
# s --> it follows log normal distribution
```

```python
processed_s = np.log(s)
```

```python
# processed_s
```

```python
# Now check whether s follows normal or gaussian distribution or not
```

```python
import pylab 
import scipy.stats as stats
```

```python
stats.probplot(processed_s, dist="norm", plot=pylab)
pylab.show()
```

```python
''' Implementation of pareto distribution '''
```

```python
lower = 0  # the lower bound for your values (Xm : scale)
shape = 40   # the distribution shape parameter, also known as `a` or `alpha`
size = 10000 # the size of your sample (number of random values)
```

```python
x = np.random.pareto(shape, size) + lower
```

```python
# x --> it follows pareto distribution
```

```python
xt,lamda = stats.boxcox(x) # it returns lambda as its second argument, which is used in second step
```

```python
# print(xt,lamda)
```

```python
stats.probplot(xt, dist="norm", plot=pylab)
pylab.show()
# This whole Q-Q plot would disturb after assigning some value to the lower(scale)
```
